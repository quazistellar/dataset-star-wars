Цель: реализовать базовый анализ датасета с помощью библиотеки Pandas на языке программирования Python.

Данный датасет называется "Персонажи из Звёздных войн", который включает в себя 87 персонажей из франшизы с числовыми данными их роста, веса, числа-даты рождения. Также в датасете есть их цвет волос, цвет кожи, цвет глаз, пол, род, транспортные средства, родной мир персонажей, фильмы (появление персонажей) и космические корабли.

Начать работу стоит с импортирования библиотеки pandas для работы с данными. После этого создаём переменную (любое удобное название)

Проведем анализ датасета с помощью функции info(), выводится следующая информация:

Проведем анализ датасета с помощью функции describe() - статистическое описание числовых данных:
- count - количество непустых значений в столбцах с числовыми значениями - name, mass, height
- mean - среднее арифметическое значение (сумма всех значений столбца, деленная на количество этих значений)
- std - стандартное отклонение относительно среднего значения, в столбце height оно небольшое -  34,7, следовательно разброс по данным небольшой, а вот в столбце mass отклонение равно 169,4 - разброс по данным существенно больше
- min показывает минимальное значение в столбце
- max аналогично показывает максимальное значение в столбце
- 25% - первый квартиль (нижний) - значение разделяет самые низкие 25% данных
- 50% - медиана - значение разделяет данные пополам
- 75% - третий квартиль (верхний) - значение разделяет самые низкие 75% данных

Заменим одни данные на другие с помощью loc:

как можно увидеть из кода ниже, данных в 77 строке по столбцу "starships" нет - выводится nan
Заменим пустые данные на "spaceship".

```python
value_one = stars.loc[77, "starships"]
value_one
```

```python
stars.loc[77, "starships"] = "spaceship"
```

Теперь в 77 строке данного столбца другие данные.

```python
value_one = stars.loc[77, "starships"]
value_one
```

Замену также можно произвести с помощью метода replace. Заменим все "nan" на "unknown"

```python
import numpy as np
stars = stars.replace({"mass":np.nan}, "unknown")
```

Посмотрим на результат:

```python
data_stars_two = stars.loc[stars["mass"] == "unknown"]
data_stars_two
```


Методом drop_duplicates() удалим дубликаты в датасете.
