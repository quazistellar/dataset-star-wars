Для начала импортируем библиотеку pandas для работы с данными, создадим переменную и присвоим ей значение для чтения csv-файла с датасетом. Выведем первые и крайние 4 его строчки.

```python
import pandas as pd

stars = pd.read_csv("star_wars_character_dataset.csv")
stars
```

Далее начнём работу непосредственно с графиками для датасета. Импортируем библиотеку matplotlib для построения графиков, а также numpy (чтобы было :D).

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

Построим линейный график для визуализации данных относительно персонажей и числовых данных их роста.

Воспользуемся библиотекой matplotlib и её методами:
- plot() - создаёт линейный график с заданными параметрами
- функция figure - это область, на которой рисуются графики, т.е создаётся новая фигура для графика
- функция figsize - задает параметры в дюймах для размера фигуры - в данном случае ширина - 30, а высота - 15
- с помощью head() библиотеки pandas берем первые 17 строчек - 17 персонажей (где указаны данные)
- label - это подписи для графиков, xlabel задает подпись по оси x, ylabel - по оси y. В легенду (то место, где находятся условные обозначения) помещаем label - рост- заголовок для обозначения роста, по оси x заголовок - имя персонажа, по оси y - значение его роста.
- colors - установление цветов для графиков
- legend() - отображение легенды с условными обозначениями для удобства
- show - отображение текущего графика

На получившемся графике наглядно видно значение роста среди выбранных персонажей - ниже всех дроид R2D2, а выше - Чубакка :D

```python
plt.figure(figsize=(30,15))
plt.plot(stars["name"].head(17), stars["height"].head(17), label = "Рост", color = "green")
plt.title("Статистика роста персонажей", color = "darkred")
plt.xlabel("Имя персонажа",  color = "darkred")
plt.ylabel("Значение роста",  color = "darkred")
plt.legend()
plt.show()
```

Создадим ещё один линейный график, но уже со статистикой возраста первых восьми персонажей: старше всех из представленных - дроид Ситрипио, младше - Лея Органа

```python
plt.figure(figsize=(20,5))
plt.plot(stars["name"].head(8), stars["birth_year"].head(8), label = "Возраст", color = "pink")
plt.title("Статистика возраста персонажей", color = "darkred")
plt.xlabel("Имя персонажа",  color = "darkred")
plt.ylabel("Значение возраста",  color = "darkred")
plt.legend()
plt.show()
```

В следующем линейном графике покажем статистику веса первых 11 персонажей из датасета: Дарту Вейдеру пора на диету, а дроидов R2D2 и R5D4 может сдуть ветром. 

```python
plt.figure(figsize=(20,10))
plt.plot(stars["name"].head(11), stars["mass"].head(11), label = "Рост", color = "blue")
plt.title("Статистика веса персонажей", color = "darkblue")
plt.xlabel("Имя персонажа",  color = "darkblue")
plt.ylabel("Значение веса",  color = "darkblue")
plt.legend()
plt.show()
```

Приступим к созданию столбчатых диаграмм, в этом поможет функция bar() - она как раз для создания столбчатых диаграмм с заданными параметрами.
Построим статистику веса персонажей, как в предыдущем линейном графике, но возьмем первых 10 персонажей. 

```python
plt.figure(figsize=(15,5))
plt.bar(stars["name"].head(10), stars["mass"].head(10), label="Вес столбцы", color = "purple")
plt.title("Статистика веса персонажей")
plt.xlabel("Имя персонажа")
plt.ylabel("Значение веса")
plt.legend()
plt.show()
```

Создадим столбчатую диаграмму для отображения статистики роста для первых 15 персонажей:

```python
plt.figure(figsize=(25,10))
plt.bar(stars["name"].head(15), stars["height"].head(15), label="Рост столбцы", color = "cyan")
plt.title("Статистика роста персонажей")
plt.xlabel("Имя персонажа")
plt.ylabel("Значение роста")
plt.legend()
plt.show()
```

Теперь используя наложение линейного графика и столбчатой диаграммы, сделаем статистику для возраста персонажей: прекрасно видно, что неизменно старше всех - мастер Йода, у двух персонажей нет данных о возрасте, а младше всех - Лея и Люк.

```python
plt.figure(figsize=(30,15))
plt.plot(stars["name"].head(20), stars["birth_year"].head(20), label="Рост", color = "orange")
plt.bar(stars["name"].head(20), stars["birth_year"].head(20), label="Рост столбцы", color = "green")
plt.title("Статистика возраста персонажей")
plt.xlabel("Имя персонажа")
plt.ylabel("Значение возраста")
plt.legend()
plt.show()
```

Создадим круговые диаграммы для соотношения количества персонажей с разными цветами глаз. Создаем новую переменную stars_ey, куда помещаем количественные данные по столбцу с цветом глаз. Далее функцией pie() создаем круговую диаграмму, передаем в заголовок индексы - названия цветов глаз, а затем, воспользуясь autopct дополняем диаграмму процентными данными. Передаем цвета и диаграмма готова: 24.1% персонажей имеют коричневые - карие глаза, 21.8% - голубые, 12.6 - желтые. В меньшинстве оказались белые, темные, зелёные (что необычно), розовые и другие нетипичные для глаз цвета.

```python
stars_ey = stars["eye_color"].value_counts()
plt.pie(stars_ey, labels=stars_ey.index, autopct="%1.1f%%", colors = ["magenta", "blue", "green", "red", "yellow", "orange", "pink", "purple", "grey", "cyan", "darkblue", "darkred", "darkgreen", "teal", "tan", "kraft", "vanilla"])
plt.title("Соотношение цвета глаз персонажей")
plt.show()
```

Сделаем ещё одну круговую диаграмму для соотношения полов. Создаем переменную и передаем туда количественные данные по столбцу с полом персонажей, добавляем процентные данные. В итоге, по диаграмме видно, что мужчин больше всех - 72.3 процента, женских персонажей - 19.3%, неуказанный пол - 7.2%, гермафродиты также оказались в наличии в нашем датасете - целых 1.2% - ожидаемо меньше всех

```python
stars_s = stars["sex"].value_counts()
plt.pie(stars_s, labels=stars_s.index, autopct="%1.1f%%", colors = ["teal", "pink", "grey", "orange"])
plt.title("Соотношение пола у персонажей")
plt.show()
```

Создадим диаграмму рассеяния. Используем для этого функцию scatter(), возьмём первые 50 персонажей и отразим связь между ростом и весом. Видим, что явной связи нет, но основные результаты сосредоточены в правой части графика.

```python
plt.figure(figsize=(10,6))
plt.scatter(stars["height"].head(50), stars["mass"].head(50), color = "purple")
plt.title("Связь ростом и весом у первых 50 персонажей")
plt.xlabel("Рост")
plt.ylabel("Вес")
plt.show()
```

Сделаем еще одну диаграмму рассеяния, но здесь уже возьмём данные роста и возраста. Явной связи также нет, но основная масса результатов сосредоточена в правой части графика - в диапазоне роста от 160 до 200, по оси веса связь почти не выходит дальше 200 

```python
plt.figure(figsize=(10,6))
plt.scatter(stars["height"].head(50), stars["birth_year"].head(50), color = "green")
plt.title("Связь между ростом и возрастом у первых 50 персонажей")
plt.xlabel("Рост")
plt.ylabel("Вес")
plt.show()
```

Далее построим гистограммы с помощью функции hist(). Bins - количество используемых прямоугольников для отображения результата, edgecolor - их цветовая обводка, alpha задает значения прозрачности гистограммы. Выберем столбец с ростом и выведем результат. Персонажи по росту в основной массе находятся в диапазоне от 150 до 230 сантиметров, что заметно по частоте и значениям роста, указанным слева и снизу соответственно. 

```python
plt.figure(figsize=(10,6))
plt.hist(stars["height"], bins=30, color = ["teal"], edgecolor = "blue", alpha=0.7, label="Распределение роста персонажей")
plt.xlabel("Значения")
plt.ylabel("Частота")
plt.legend()
plt.show()
```

Построим гистограмму для столбца с весом персонажей, в основном их диапазон от приблизительно 0 до 200 кг, но на гистограмме в правом нижнем углу видно персонажа из датасета, чей вес равен 1358 кг, что явно выбивается из общей картины xD

```python
plt.figure(figsize=(10,6))
plt.hist(stars["mass"], bins=40, color = ["pink"], edgecolor = "red", alpha=0.7, label="Распределение веса персонажей")
plt.legend()
plt.show()
```

Сделаем гистограмму для распределения данных по возрасту персонажей, здесь результаты разные: видим старейших персонажей ближе к правой части экрана, но основная масса в диапазоне от 0 до 200

```python
plt.figure(figsize=(10,6))
plt.hist(stars["birth_year"], bins=40, color = ["yellow"], edgecolor = "orange", alpha=0.7, label="Распределение возраста персонажей")
plt.legend()
plt.show()
```
